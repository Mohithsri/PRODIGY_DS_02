```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset from a URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Explore the data
print(df.head())  # Display the first few rows of the dataset
print(df.shape)  # Check the dimensions of the dataset
print(df.describe())  # Get summary statistics of numerical variables
print(df.dtypes)  # Check the data types of variables
print(df.isnull().sum())  # Check for missing values

# Data cleaning
# Fill missing values in the 'Embarked' column with the mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Remove rows with missing values in the 'Age' column
df.dropna(subset=['Age'], inplace=True)

# Calculate the median age
median_age = df['Age'].median()

# Fill missing values in the 'Age' column with the median age
df['Age'] = df['Age'].fillna(median_age)

# Data visualization
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()

sns.histplot(df, x='Survived', hue='Sex', multiple='stack', bins=2)
plt.xlabel('Survival Status')
plt.ylabel('Frequency')
plt.title('Survival Count by Sex')
plt.show()

sns.histplot(df, x='Survived', hue='Pclass', multiple='stack', bins=2)
plt.xlabel('Survival Status')
plt.ylabel('Frequency')
plt.title('Survival Count by Passenger Class')
plt.show()

# Explore relationships
correlation = df[['Age', 'Fare']].corr()
print(correlation)

cross_tab = pd.crosstab(df['Pclass'], df['Survived'])
print(cross_tab)
```
